```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]
img = cv2.imread(filename)
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(img, (100,50))
cv2_imshow(resized)
```

```python
resized = cv2.resize(gray, (100,50))
cv2_imshow(resized)
```

```python
a = [img,gray]
for i in a:
  resized=cv2.resize(i,(100,50))
  cv2_imshow(resized)
```

```python
anno = img.copy()
cv2.rectangle(anno, (10,100),(700,700),(251,250,50), 5)
cv2.putText(anno,'My Box',(60,45),cv2.FONT_HERSHEY_SIMPLEX,1,(200,0,200), 5)
cv2_imshow(anno)
```

```python
cv2.imwrite('output_img.jpg',anno)
print("Image saved as output_img.jpg")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
img = load_img(filename,target_size =(244,244))
img_array = img_to_array(img)
img_array = np.expand_dims(img_array,axis=0)
```

```python
# Define the augmentation generator
datagen = ImageDataGenerator(
    rotation_range=30,
    #width_shift_range=0.2,
    #height_shift_range=0.2,
    #shear_range=0.2,
    #zoom_range=0.2,
    #horizontal_flip=True,
    #brightness_range = [0.5,1.5]'''
)
# Generate augmented images
i = 0
for batch in datagen.flow(img_array, batch_size = 1):
  plt.figure(figsize = (2,2))
  plt.imshow(batch[0].astype('uint8'))
  plt.axis('off')
  i += 1
  if i > 5:
    break
```
